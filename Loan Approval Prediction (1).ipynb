```python
inputs = df.drop('Loan_Status', axis = 1)
inputs
```

```python
target = df['Loan_Status']
target
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
inputs['Gender_n'] = lb.fit_transform(inputs['Gender'])
inputs['Married_n'] = lb.fit_transform(inputs['Married'])
inputs['Education_n'] = lb.fit_transform(inputs['Education'])
inputs['Self_Employed_n'] = lb.fit_transform(inputs['Self_Employed'])
inputs['Property_Area_n'] = lb.fit_transform(inputs['Property_Area'])
inputs['Dependents_n'] = lb.fit_transform(inputs['Dependents'])
inputs_n = inputs.drop(['Gender','Married','Education','Self_Employed','Property_Area','Dependents'], axis = 1)
inputs_n
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs_n, target, test_size = 0.33, random_state = 0)
```

```python
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
